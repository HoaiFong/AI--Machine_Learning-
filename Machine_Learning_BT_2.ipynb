```python
import pandas as pd
import numpy as np
url = 'lung_cancer_examples.csv'
data = pd.read_csv(url)
from matplotlib import pyplot as plt
data.hist() # h√†m histogram
plt.show()
```

```python
correlation = data.corr() #m·ªëi t∆∞∆°ng quan gi·ªØa c√°c gi√° tr·ªã 
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter',x = 'Alkhol',y='Smokes',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Alkhol']),y= data['Smokes'])
print(model.intercept_)
print(model.coef_)
# y = coef*x + intercept
# coef : h·ªá s·ªë g√≥c, intercept : giao ƒëi·ªÉm v·ªõi tr·ª• Y
model.score(X = pd.DataFrame(data['Alkhol']),y= data['Smokes']) #ƒë·ªô ch√≠nh x√°c
```

```python
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Smokes', figsize = (9,9), color = 'black', xlim = (0.7))
y_pred = model.predict(X = pd.DataFrame(data['Alkhol']))
plt.plot(data['Alkhol'],y_pred,color='blue')
```

```python
# h√†m b·∫≠c 3
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Alkhol'], data['Alkhol']**2,data['Alkhol']**3]).T
model.fit(X, y=data['Smokes'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
z = model.score(X,y =data['Smokes'])
print('Do chinh xac cua mo hinh: ',z)
```

```python
range = np.arange(1,9,0.1)
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Smokes', figsize = (9,9), color = 'black', xlim = (0.7))
predictor = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
```

```python
# h√†m b·∫≠c 5
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Alkhol'], data['Alkhol']**2,data['Alkhol']**3,data['Alkhol']**4,data['Alkhol']**5]).T
model.fit(X, y=data['Smokes'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
z = model.score(X,y =data['Smokes'])
print('Do chinh xac cua mo hinh: ',z)
```
