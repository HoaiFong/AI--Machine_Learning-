```python
import pandas as pd
import numpy as np
url = 'iris.csv'
data = pd.read_csv(url)
from matplotlib import pyplot as plt
data.hist() # h√†m histogram
plt.show()
```

```python
correlation = data.corr() #m·ªëi t∆∞∆°ng quan gi·ªØa c√°c gi√° tr·ªã 
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter',x = 'petal.width',y='petal.length',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['petal.width']),y= data['petal.length'])
print(model.intercept_)
print(model.coef_)
# y = coef*x + intercept
# coef : h·ªá s·ªë g√≥c, intercept : giao ƒëi·ªÉm v·ªõi tr·ª• Y
model.score(X = pd.DataFrame(data['petal.width']),y= data['petal.length']) #ƒë·ªô ch√≠nh x√°c
```

```python
data.plot(kind = 'scatter', x = 'petal.width', y = 'petal.length', figsize = (9,9), color = 'black')
y_pred = model.predict(X = pd.DataFrame(data['petal.width']))
plt.plot(data['petal.width'],y_pred,color='blue')
```

```python
# h√†m b·∫≠c 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['petal.width'], data['petal.width']**2]).T
model.fit(X, y=data['petal.length'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
z = model.score(X,y =data['petal.length'])
print('Do chinh xac cua mo hinh: ',z)
```

```python
range = np.arange(0,3,0.1)
data.plot(kind = 'scatter', x = 'petal.width', y = 'petal.length', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
```

```python
# h√†m b·∫≠c 5
model = linear_model.LinearRegression()
X = pd.DataFrame([data['petal.width'], data['petal.width']**2,data['petal.width']**3,data['petal.width']**4,data['petal.width']**5]).T
model.fit(X, y=data['petal.length'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
z = model.score(X,y =data['petal.length'])
print('Do chinh xac cua mo hinh: ',z)
```

```python
range = np.arange(0,3,0.1)
data.plot(kind = 'scatter', x = 'petal.width', y = 'petal.length', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
```
