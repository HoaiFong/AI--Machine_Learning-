```python
import pandas as pd
import numpy as np
url = 'crabs.csv'
data = pd.read_csv(url)
from matplotlib import pyplot as plt
data.hist() # h√†m histogram
plt.show()
```

```python
correlation = data.corr() #m·ªëi t∆∞∆°ng quan gi·ªØa c√°c gi√° tr·ªã 
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter',x = 'depth',y='fontallip',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['depth']),y= data['fontallip'])
print(model.intercept_)
print(model.coef_)
# y = coef*x + intercept
# coef : h·ªá s·ªë g√≥c, intercept : giao ƒëi·ªÉm v·ªõi tr·ª• Y
model.score(X = pd.DataFrame(data['depth']),y= data['fontallip']) #ƒë·ªô ch√≠nh x√°c
```

```python
data.plot(kind = 'scatter', x = 'depth', y = 'fontallip', figsize = (9,9), color = 'black', xlim = (0.7))
y_pred = model.predict(X = pd.DataFrame(data['depth']))
plt.plot(data['depth'],y_pred,color='blue')
```

```python
# h√†m b·∫≠c 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['depth'], data['depth']**2]).T
model.fit(X, y=data['fontallip'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
z = model.score(X,y =data['fontallip'])
print('Do chinh xac cua mo hinh: ',z)
```

```python
range = np.arange(2.5,30,0.1)
data.plot(kind = 'scatter', x = 'depth', y = 'fontallip', figsize = (9,9), color = 'black', xlim = (0.7))
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
```

```python
# h√†m b·∫≠c 5
model = linear_model.LinearRegression()
X = pd.DataFrame([data['depth'], data['depth']**2,data['depth']**3,data['depth']**4,data['depth']**5]).T
model.fit(X, y=data['fontallip'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
z = model.score(X,y =data['depth'])
print('Do chinh xac cua mo hinh: ',z)
```

```python
range = np.arange(2.5,25,0.1)
data.plot(kind = 'scatter', x = 'depth', y = 'fontallip', figsize = (9,9), color = 'black', xlim = (0.7))
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
```
